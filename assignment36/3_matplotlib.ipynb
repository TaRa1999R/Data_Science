```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
x = np.array ([0 , 10 , 20])
y = np.array ([0 , 20 , 10])
```

```python
plt.plot (x , y)
plt.show ()
```

```python
plt.subplot (3 , 2 , 1)
plt.plot (x , y , color = 'red')
plt.xlabel ('x points')
plt.ylabel ('y points')
plt.title ('red line')

plt.subplot (3 , 2 , 2)
plt.plot (x , y , '-.' ,color = 'c')
plt.xlabel ('x points')
plt.ylabel ('y points')
plt.title ('cyan dashed-dotted line')

plt.subplot (3 , 2 , 3)
plt.plot (x , y , 'o' , )
plt.xlabel ('x points')
plt.ylabel ('y points')
plt.title ('red line')

plt.subplot (3 , 2 , 4)
plt.plot (x , y , color = 'red')
plt.xlabel ('x points')
plt.ylabel ('y points')
plt.title ('red line')

plt.subplot (3 , 2 , 5)
plt.plot (x , y , color = 'red')
plt.xlabel ('x points')
plt.ylabel ('y points')
plt.title ('red line')

plt.subplot (3 , 2 , 6)
plt.plot (x , y , color = 'red')
plt.xlabel ('x points')
plt.ylabel ('y points')
plt.title ('red line')

plt.suptitle ('PLOTS')
plt.show ()
```
