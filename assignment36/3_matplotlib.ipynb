```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
x = np.array ([0 , 10 , 20])
y = np.array ([0 , 20 , 10])
```

```python
plt.plot (x , y , c = 'red')
plt.show ()
```

```python
plt.plot (x , y , 'o')
plt.show ()
```

```python
plt.plot (x , y , marker = 'o')
plt.show ()
```
