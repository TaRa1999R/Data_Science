```python
import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd
```

```python
data = pd.read_csv ('inputs\Lionel Messi Goals.csv')
data.head (10)
```

Pre-Processing on Minute

```python
def fix_min (min) :
    return eval (min)

data['Minute'] = data ['Minute'].apply (fix_min)
data.head (6)
```

Pre-Processing on Date to Produce Year

```python
def find_year (date) :
    year  = date.split()
    return year [3]

data ['Year'] = data['Date'].apply (find_year)
data.head(6)
```

Pre-Processing on Columns Titel

```python
data = data.rename (columns = { 'Goal Type':'Goal_Type' 
                              , 'Goal Method':'Goal_Method' 
                              , 'Score Home Team - When Messi Scored' : 'Score_Home_Team'
                              , 'Score Away Team - When Messi Scored' : 'Score_Away_Team'})
data.head(3)
```

Messi's Goals in Each Year

```python
df_year = data ['Year'].value_counts ().reset_index ().sort_values (by = ['Year'] , ascending = True)
print (df_year)

plt.figure (figsize = (14 , 8))
plt.bar (df_year['Year'] , df_year['count'] , color = 'c')

plt.xlabel ('Year')
plt.ylabel ('Number of Goals')
plt.title ("Messi's Goals in Each Year")
plt.grid (axis= 'y')

plt.savefig ('outputs\output2 goals in each year.png')
plt.show ()
```

Are most of the penalties with the right foot or the left foot?

```python
df_penalty = data.query ('Goal_Type == "Penalty"')
df_foot = df_penalty['Goal_Method'].value_counts().reset_index()
right = pd.DataFrame ([{'Goal_Method':'Right foot' , 'count': '0'}])
df_foot = pd.concat ([df_foot ,right]).reset_index(drop = True)
print (df_foot)

plt.figure (figsize = (7 , 7))
plt.pie (df_foot['count'] , labels = df_foot['Goal_Method'] , colors = 'cyan' , autopct='%1.1f%%', startangle=180)

plt.title ("Right foot Vs. Left foot in Penalties")
plt.legend ()

plt.savefig ('outputs\output2 Right foot Vs. Left foot.png')
plt.show ()
```

What percentage of the games in which Messi scored in overtime ended in victory?

```python
df_overtime = data.query ('Minute > 90')
df_result = df_overtime['Final Game Result'].value_counts().reset_index()
print (df_result)

plt.figure (figsize = (7 , 7))
plt.pie ( df_result['count'] , labels = df_result['Final Game Result'] ,colors=['c', 'y', 'm'] 
        , autopct='%1.1f%%', startangle = 180 , explode = (0.2 , 0 , 0))

plt.title ("Game Results when Messi Scored in overtime")
plt.legend ()

plt.savefig ('outputs\output2 victory by overtime goals.png')
plt.show ()
```

What percentage of Masi's goals were draws ?

```python
draw = data.query ('Score_Home_Team == Score_Away_Team')
df_draw = pd.DataFrame ([{'Result' : 'Draw' , 'Count' : len(draw)}
                     ,{'Result' : 'UnDraw' , 'Count' : len(data) - len(draw)}])
print (df_draw)

plt.figure (figsize = (7 , 7))
plt.pie ( df_draw['Count'] , labels = df_draw['Result'] ,colors=['m', 'c'] 
        , autopct='%1.1f%%', startangle = 180 , explode = (0.1 , 0))

plt.title ("Result in Draw or Not")
plt.legend ()

plt.savefig ('outputs\output2 draw or not.png')
plt.show ()
```
