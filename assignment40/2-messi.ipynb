```python
import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd
```

```python
data = pd.read_csv ('inputs\Lionel Messi Goals.csv')
data.head (10)
```

Pre-Processing on Minute

```python
def fix_min (min) :
    return eval (min)

data['Minute'] = data ['Minute'].apply (fix_min)
data.head (6)
```

Pre-Processing on Date to Produce Year

```python
def find_year (date) :
    year  = date.split()
    return year [3]

data ['Year'] = data['Date'].apply (find_year)
data.head(10)
```

Messi's Goals in Each Year

```python
df_year = data ['Year'].value_counts ().reset_index ().sort_values (by = ['Year'] , ascending = True)
print (df_year)

plt.figure (figsize = (20 , 18))
plt.bar (df_year['Year'] , df_year['count'])

plt.xlabel ('Year')
plt.ylabel ('Number of Goals')
plt.title ("Messi's Goals in Each Year")
plt.grid (axis= 'y')

plt.savefig ('outputs\output2 goals in each year.png')
plt.show ()
```
