```python
import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd
```

```python
data = pd.read_csv ('inputs\Lionel Messi Goals.csv')
data.head (10)
```

Pre-Processing on Minute

```python
def fix_min (min) :
    return eval (min)

data['Minute'] = data ['Minute'].apply (fix_min)
data.head (6)
```

Pre-Processing on Date to Produce Year

```python
def find_year (date) :
    year  = date.split()
    return year [3]

data ['Year'] = data['Date'].apply (find_year)
data.head(6)
```

Pre-Processing on Columns Titel

```python
data = data.rename (columns = {'Goal Type':'Goal_Type' , 'Goal Method':'Goal_Method'})
data.head(3)
```

Messi's Goals in Each Year

```python
df_year = data ['Year'].value_counts ().reset_index ().sort_values (by = ['Year'] , ascending = True)
print (df_year)

plt.figure (figsize = (14 , 8))
plt.bar (df_year['Year'] , df_year['count'] , color = 'green')

plt.xlabel ('Year')
plt.ylabel ('Number of Goals')
plt.title ("Messi's Goals in Each Year")
plt.grid (axis= 'y')

plt.savefig ('outputs\output2 goals in each year.png')
plt.show ()
```

Are most of the penalties with the right foot or the left foot?

```python
df_penalty = data.query ('Goal_Type == "Penalty"')
df_foot = df_penalty['Goal_Method'].value_counts().reset_index()
right = pd.DataFrame ([{'Goal_Method':'Right foot' , 'count': '0'}])
df_foot = pd.concat ([df_foot ,right]).reset_index(drop = True)
print (df_foot)

plt.figure (figsize = (10 , 10))
plt.pie (df_foot['count'] , labels = df_foot['Goal_Method'] , colors = 'green')

plt.title ("Right foot Vs. Left foot in Penalties")
plt.legend ()

plt.savefig ('outputs\output2 Right foot Vs. Left foot.png')
plt.show ()
```
