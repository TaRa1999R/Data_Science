```python
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd
```

```python
data = pd.read_csv ('inputs\covid19_fullgrouped.csv')
data.head (10)
```

```python
data ['Date'] = pd.to_datetime (data['Date'])
data ['Month'] = data['Date'].map (lambda date : date.month)
data ['Day'] = data['Date'].map (lambda date : date.day)
data.head (10)
```

Fourth Month Corona Patients

```python
fourth = data.query ('Month == 4')
month = fourth.groupby ('Country')['Active'].max().reset_index()
sorted = month.sort_values ('Active' , ascending = False).reset_index(drop = True)
sorted.head (8)
```

Death Rate In Iran

```python
iran_fourth = data.query ('Country == "Iran" and Month == 4') 
iran_fourth.head(5)

plt.figure (figsize = (12 , 6))
plt.plot ( iran_fourth ['Day'].to_numpy ()
         , iran_fourth ['New cases'].to_numpy()
         , color = 'skyblue'
         , label = 'cases')
plt.plot ( iran_fourth ['Day'].to_numpy ()
         , iran_fourth ['New deaths'].to_numpy()
         , color = 'gold'
         , label = 'deaths')

plt.title ('Death Rate in Iran in Fourth-Month Day ')
plt.xlabel ('days')
plt.ylabel ('number of people')
plt.legend ()
plt.savefig ('outputs\output_2_death rate iran.png')
plt.plot ()
```
