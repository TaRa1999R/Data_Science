```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
balloons = pd.DataFrame ({
    'width' : np.random.normal (30 , 0.5 , 500) ,   #cm
    'length' : np.random.normal (30 , 0.5 , 500) ,  #cm
    'weight' : np.random.normal (0.015 , 0.2 , 500) ,
    'class' : np.ones (500 , dtype='int') })


melons = pd.DataFrame ({
    'width' : np.random.normal (30 , 0.5 , 500) ,   #cm
    'length' : np.random.normal (30 , 0.5 , 500) ,  #cm
    'weight' : np.random.normal (3 , 0.5 , 500) ,   #Kg
    'class' : np.zeros (500 , dtype='int') })

items = pd.concat([balloons, melons])
```

```python
fig = plt.figure()
ax = fig.add_subplot( 111 , projection='3d')

ax.scatter (balloons['length'] , balloons['width'] , balloons['weight'] , c = 'red' , marker = '*' , alpha = 0.3 , label = 'Balloons')
ax.scatter (melons['length'] , melons['width'] , melons['weight'] , c = 'green' , marker = 'D' , alpha = 0.3 , label = 'Melons')

ax.set_xlabel ('Length$_{(cm)}$')
ax.set_ylabel ('Width$_{(cm)}$')
ax.set_zlabel ('Weight$_{(Kg)}$')
ax.legend ()

plt.title ("Ballons Vs. Melons")
plt.savefig ("outputs\Balloons & Melons.png")
plt.show ()
```
