```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
```

```python
db = pd.read_csv ("inputs\california_cities.csv")
db.head ()
```

```python
plt.scatter ( db ['latd'] 
            , db ['longd']
            , s = db ['area_total_km2']
            , alpha = 0.5
            , linewidths = 0
            , c = np.log10 (db ['population_total'])
            , cmap = 'gist_ncar' )

plt.colorbar (label = 'Log$_{10}$ (population)')
plt.xlabel ('Latitude')
plt.ylabel ('Longtitude')
plt.title ('Area and Population of California cities')

plt.grid ()
plt.legend ()

plt.savefig ('outputs\california_cities.png')
plt.show ()
```
