```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
```

```python
db = pd.read_csv ("inputs\california_cities.csv")
db.head ()
```

```python
plt.scatter ( db['latd']
            , db['longd']
            , s = db['area_total_km2']
            , alpha = 0.5
            , linewidths = 0
            , c = np.log10(db ['population_total'])
            , cmap = 'gist_ncar')

plt.colorbar (label = 'log$_{10}$(population)')
plt.xlabel ('Latitude')
plt.ylabel ('Longtitude')
plt.title ('Area and Population of California cities')

plt.clim (3 , 7)
for area in [100, 300, 500]:
    plt.scatter([] , [] , c='k' , alpha=0.3 , s=area , label = str(area) + 'km$^2$')
plt.legend( scatterpoints = 1
          , frameon = False
          , labelspacing = 1
          , title = 'City Areas' )
plt.grid ()

plt.savefig ('outputs\California cities.png')
plt.show ()
```
