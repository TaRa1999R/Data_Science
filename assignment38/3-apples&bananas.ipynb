```python
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd
```

```python
n = 1000
std = 0.6
apples = pd.DataFrame ({
    'width' : np.random.normal (6 , std , n) ,
    'length' : np.random.normal (6 , std , n) ,
    'class' : np.ones (n , dtype='int') })


bananas = pd.DataFrame ({
    'width' :  np.random.normal (4 , std , n) ,
    'length' : np.random.normal (8 , std , n) ,
    'class' : np.zeros (n , dtype='int') })

fruits = pd.concat([bananas, apples])
```

```python
plt.scatter (apples['length'] , apples['width'] , label = 'Apples' , color = 'red' , alpha = 0.5 , linewidths = 0)
plt.scatter (bananas['length'] , bananas['width'] , label = 'Banana' , color = 'gold' , alpha = 0.5 , linewidths = 0)

plt.legend ()
plt.title ('Bananas & Apples distribution on scatter plot')
plt.xlabel ('Length')
plt.ylabel ('Width')

plt.savefig ("outputs\Banana & Apple Scatter plot.png")
plt.show ()
```

```python
labels = ['bananas' , 'apples' , 'bananas' , 'apples']

x = [1 , 1.5 , 2.5 , 3.0]

height = [ fruits.loc[fruits['class'] == 0 , 'length'].mean()
         , fruits.loc[fruits['class'] == 1 , 'length'].mean()
         , fruits.loc[fruits['class'] == 0 , 'width'].mean()
         , fruits.loc[fruits['class'] == 1 , 'width'].mean() ]

yerr = [ 2*fruits.loc[fruits['class'] == 0 , 'length'].std()
       , 2*fruits.loc[fruits['class'] == 1 , 'length'].std()
       , 2*fruits.loc[fruits['class'] == 0 , 'width'].std() 
       , 2*fruits.loc[fruits['class'] == 1 , 'width'].std() ]

color = ['tab:blue', 'tab:blue', 'tab:orange', 'tab:orange']

plt.bar ( x = x
        , height=height
        , color = color
        , width = 0.4
        , error_kw = {'lw': 4}
        , yerr = yerr )

plt.xticks(ticks = x , labels = labels)
plt.title('Fruit length VS width')

plt.grid()

plt.savefig ("outputs\Banana & Apple Bar chart.png")
plt.show ()
```

```python
fea = 'length'

plt.boxplot ( x = fea
            , data = fruits.loc[fruits['class'] == 0]
            , patch_artist = True 
            , boxprops = {'facecolor': 'tab:blue'}
            , positions = [0]
            , labels = ['bananas']
            , widths = [0.15] )

plt.boxplot ( x = fea
            , data = fruits.loc[fruits['class'] == 1]
            , patch_artist = True
            , boxprops = {'facecolor': 'tab:orange'}
            , positions = [0.3]
            , labels = ['apples']
            , widths = [0.15] )

plt.ylabel('length')
plt.xlim(-0.2, 0.5)
plt.title('Length Distribution ')

plt.savefig ("outputs\Banana & Apple Box diagram.png")
plt.show ()
```
