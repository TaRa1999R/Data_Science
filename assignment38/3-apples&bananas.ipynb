```python
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd
```

```python
apple_w = np.random.normal (6 , 0.6 , 1000)
apple_l = np.random.normal (6 , 0.6 , 1000)

banana_w = np.random.normal (4 , 0.6 , 1000)
banana_l = np.random.normal (8 , 0.6 , 1000)
```

```python
plt.scatter (apple_w , apple_l , label = 'Apples' , color = 'red')
plt.scatter (banana_w , banana_l , label = 'Banana' , color = 'yellow')

plt.legend ()
plt.title ('Bananas & Apples distribution on scatter plot')
plt.xlabel ('Length')
plt.ylabel ('Width')

plt.savefig ("outputs\Banana & Apple Scatter plot.png")
plt.show ()
```
