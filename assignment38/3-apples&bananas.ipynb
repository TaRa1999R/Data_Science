```python
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd
```

```python
apples = pd.DataFrame ({
    'width' : np.random.normal (6 , 0.6 , 1000) ,
    'length' : np.random.normal (6 , 0.6 , 1000) ,
    'class' : np.ones (1000 , dtype='int') })


bananas = pd.DataFrame ({
    'width' :  np.random.normal (4 , 0.6 , 1000) ,
    'length' : np.random.normal (8 , 0.6 , 1000) ,
    'class' : np.zeros (1000 , dtype='int') })

fruits = pd.concat([bananas, apples])
```

```python
plt.scatter (apples['length'] , apples['width'] , label = 'Apples' , color = 'red' , alpha = 0.5 , linewidths = 0)
plt.scatter (bananas['length'] , bananas['width'] , label = 'Banana' , color = 'yellow' , alpha = 0.5 , linewidths = 0)

plt.legend ()
plt.title ('Bananas & Apples distribution on scatter plot')
plt.xlabel ('Length')
plt.ylabel ('Width')

plt.savefig ("outputs\Banana & Apple Scatter plot.png")
plt.show ()
```

```python
labels = ['bananas-L' , 'apples-L' , 'bananas-W' , 'apples-W']

x = [1 , 1.5 , 2.5 , 3.0]

height = [ fruits.loc[fruits['class'] == 0 , 'length'].mean()
         , fruits.loc[fruits['class'] == 1 , 'length'].mean()
         , fruits.loc[fruits['class'] == 0 , 'width'].mean()
         , fruits.loc[fruits['class'] == 1 , 'width'].mean() ]

yerr = [ 2*fruits.loc[fruits['class'] == 0 , 'length'].std()
       , 2*fruits.loc[fruits['class'] == 1 , 'length'].std()
       , 2*fruits.loc[fruits['class'] == 0 , 'width'].std() 
       , 2*fruits.loc[fruits['class'] == 1 , 'width'].std() ]

color = ['gold', 'red', 'gold', 'red']

plt.bar ( x = x
        , height=height
        , color = color
        , width = 0.4
        , error_kw = {'lw': 4}
        , yerr = yerr )

plt.xticks(ticks = x , labels = labels)
plt.title('Fruit length VS width')

plt.grid()

plt.savefig ("outputs\Banana & Apple Bar chart.png")
plt.show ()
```

```python
fea = 'length'

plt.boxplot ( x = fea
            , data = fruits.loc[fruits['class'] == 0]
            , patch_artist = True 
            , boxprops = {'facecolor': 'gold'}
            , positions = [0]
            , labels = ['bananas']
            , widths = [0.15] )

plt.boxplot ( x = fea
            , data = fruits.loc[fruits['class'] == 1]
            , patch_artist = True
            , boxprops = {'facecolor': 'red'}
            , positions = [0.3]
            , labels = ['apples']
            , widths = [0.15] )

plt.ylabel('length')
plt.xlim(-0.2, 0.5)
plt.title('Length Distribution ')

plt.savefig ("outputs\Banana & Apple Box diagram.png")
plt.show ()
```

```python
fea = 'width'

plt.hist ( x = fea
         , data = fruits.loc[fruits['class'] == 0]
         , bins = 30
         , density = True
         , color = 'gold' 
         , label='bananas')

plt.hist ( x = fea
         , data = fruits.loc[fruits['class'] == 1]
         , bins = 30
         , density = True
         , color = 'red'
         , label = 'apples'
         , alpha = 0.8)

plt.legend (frameon = False)
plt.title ('Width Distribution ')
plt.xlabel ('Width')

plt.savefig ("outputs\Banana & Apple Histogram.png")
plt.show ()
```

```python
fig, (ax, cax) = plt.subplots (1, 2, gridspec_kw={'width_ratios': [0.9, 0.1]})
_, _, _, mappable = ax.hist2d ( x = 'length'
                              , y = 'width'
                              , bins = 50
                              , data = fruits
                              , cmap = plt.cm.hot)
ax.set_xlabel('length')
ax.set_ylabel('width')

plt.colorbar( mappable = mappable , cax = cax , ax = ax)

plt.savefig ("outputs\Banana & Apple joint distribution histogram.png")
plt.show ()
```

```python
fea = 'L/W'
fruits[fea] = fruits['length'] / fruits['width']

plt.hist ( x = fea
         , data = fruits.loc[fruits['class'] == 0]
         , bins = 30
         , density = True
         , color = 'gold'
         , label = 'bananas' )

plt.hist ( x = fea
         , data = fruits.loc[fruits['class'] == 1]
         , bins = 30
         , density = True
         , color = 'red'
         , label = 'apples' )

plt.legend()
plt.title('L/W Distribution ')
plt.xlabel('L/W ratio')

plt.savefig ("outputs\Banana & Apple Aspect ratio.png")
plt.show ()
```

```python
fea = 'area'
fruits[fea] = fruits['length'] * fruits['width']

plt.hist ( x = fea 
         , data = fruits.loc[fruits['class'] == 0]
         , histtype = 'step'
         , lw = 4 
         , bins = 30
         , density = True
         , color = 'gold'
         , label = 'bananas')

plt.hist ( x = fea 
         , data = fruits.loc[fruits['class'] == 1]
         , histtype = 'step'
         , lw = 4
         , bins = 30
         , density = True
         , color = 'red'
         , label = 'apples' )

plt.title('Area Distribution ')
plt.legend()
plt.xlabel('Area')

plt.savefig ("outputs\Banana & Apple measure of area.png")
plt.show ()
```

```python
fea = 'circumference'
fruits[fea] = 2 * (fruits['length'] + fruits['width'])

plt.hist ( x = fea
         , data = fruits.loc[fruits['class'] == 0]
         , histtype = 'step'
         , lw = 4
         , bins = 30
         , density = True
         , color = 'gold'
         , label = 'bananas')

plt.hist ( x = fea
         , data = fruits.loc[fruits['class'] == 1]
         , histtype = 'step'
         , lw = 4 
         , bins = 30
         , density = True
         , color = 'red'
         , label = 'apples')

plt.title('circumference Distribution ')
plt.legend()
plt.xlabel('Circumference')

plt.savefig ("outputs\Banana & Apple perimeter.png")
plt.show ()
```
